```python
from sklearn.model_selection import train_test_split

x = data.drop(['SALE_PRC'], axis = 1)
y = data['SALE_PRC']
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
train_data = x_train.join(y_train)
```

```python
train_data
```

```python
train_data.hist(figsize=(15,8))
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True, cmap="YlGnBu")
```

```python
train_data['PARCELNO'] = np.log(train_data['PARCELNO']+1)
train_data['LND-SQFOOT'] = np.log(train_data['LND_SQFOOT']+1)
train_data['TOT_LVG_AREA'] = np.log(train_data['TOT_LVG_AREA']+1)
train_data['SPEC_FEAT_VAL'] = np.log(train_data['SPEC_FEAT_VAL']+1)
```

```python
train_data.hist(figsize=(15,8))
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True, cmap="YlGnBu")
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x="LATITUDE", y="LONGITUDE", data=train_data, hue="SALE_PRC", palette="coolwarm")
```

```python
from sklearn.linear_model import LinearRegression

x_train = train_data.drop(['SALE_PRC'], axis=1)
y_train = train_data['SALE_PRC']

reg = LinearRegression()

reg.fit(x_train, y_train)
```

```python
test_data = x_test.join(y_test)

test_data['PARCELNO'] = np.log(test_data['PARCELNO']+1)
test_data['LND-SQFOOT'] = np.log(test_data['LND_SQFOOT']+1)
test_data['TOT_LVG_AREA'] = np.log(test_data['TOT_LVG_AREA']+1)
test_data['SPEC_FEAT_VAL'] = np.log(test_data['SPEC_FEAT_VAL']+1)

x_test = test_data.drop(['SALE_PRC'], axis=1)
y_test = test_data['SALE_PRC']

test_data
```

```python
reg.score(x_test,y_test)
```
